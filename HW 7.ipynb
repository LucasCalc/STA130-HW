## 1. Explain succinctly in your own words (but working with a ChatBot if needed)...

1. the difference between Simple Linear Regression and Multiple Linear Regression; and the benefit the latter provides over the former

2. the difference between using a continuous variable and an indicator variable in Simple Linear Regression; and these two linear forms

3. the change that happens in the behavior of the model (i.e., the expected nature of the data it models) when a single indicator variable is introduced alongside a continuous variable to create a Multiple Linear Regression; and these two linear forms (i.e., the Simple Linear Regression versus the Multiple Linear Regression)

4. the effect of adding an interaction between a continuous and an indicator variable in Multiple Linear Regression models; and this linear form

5. the behavior of a Multiple Linear Regression model (i.e., the expected nature of the data it models) based only on indicator variables derived from a non-binary categorical variable; this linear form; and the necessarily resulting binary variable encodings it utilizes

### My Answers

1. 

Simple Linear Regression involves one independant variable and a dependant variable:
#### <center> $Y_{i}$ = $β_{0}$ + $β_{1}$$x_{i}$ + $ε_{i}$ where $ε_{i}$ = N(0,$σ^{2}$) </center>
On the other hand, Mulitple Linear Regression involves two or more independant variables.
#### <center> $Y_{i}$ = $β_{0}$ + $β_{1}$$x_{1}$ + $β_{2}$$x_{2}$ + ... + $β_{n}$$x_{n}$ + $ε_{i}$ where $ε_{i}$ = N(0,$σ^{2}$) </center>
The benefits over SLR that MLR has is that it accounts for multiple independant variables. This creates a more "detailed" relationship between factors and an outcome. Furthermore, this means that the outcome is more accurately predicted, as more factors are accounted for. 

2.

In SLR, an indicator variable is used to represent categories of the data, therefore making the relationship represent some difference between these categories. The equation looks like
#### <center> $Y_{i}$ = $β_{0}$ + $β_{1}$$1_{"category 1"}$$k_{i}$ + $ε_{i}$ where $ε_{i}$ = N(0,$σ^{2}$) </center>
The key difference in this notation is the use of $1_{"category 1"}$$k_{i}$ instead of $x_{i}$. In an example where we are comparing two categories, called category 1 and 2, each category will be assigned a number. In the above equation, category 1 is assigned the number 1, meaning category 2 is assigned 0. Depending on which category the data point is in, $k_{i}$ will take the corresponding number. Therefore, $β_{1}$ represents the contrast of the measured statistic, between the two categories.

3. 

If you have MLR with one continuous and one indicator variable, the model will represent both the effect of the continuous variable, as well as the contrast between groups of the indicator variable. 
#### <center> $Y_{i}$ = $β_{0}$ + $β_{1}$$x_{1}$+ $β_{2}$$1_{"category 1"}$$k_{2}$ + $ε_{i}$ where $ε_{i}$ = N(0,$σ^{2}$) </center>
The effects of the continuous variable will be the same between the categories. Then, the categorical difference will be applied afterwards. With this form of model, we can view the linear relationship between the continous and the outcome, while also seeing the relationship within the categories.

4. 

Adding an interaction between the indicator variable and the continous variable would change the equation to this form:
#### <center> $Y_{i}$ = $β_{0}$ + $β_{1}$$x_{1}$ + $β_{2}$$1_{"category 1"}$$k_{2}$ + $β_{3}$($x_{1}$ * $k_{2}$) + $ε_{i}$ where $ε_{i}$ = N(0,$σ^{2}$) </center>
In this equation, as previously described, one category will be represented with a k value of 0, and the other as 1. Therefore, if the data is from the category with k=0, the new term with $β_{3}$ will all become 0. On the other hand, if the data is from the category for which k=1, the data point $x_{1}$ will be mulitplied by the interaction factor $β_{3}$. Therefore, this parameter $β_{3}$ represents the contrast in the relationship between an $x_{1}$ in one category, as opposed to the other.

5.

In MLR with only indicicator variables, the equation would look as follows:
#### <center> $Y_{i}$ = $β_{0}$ + $β_{1}$$1_{"category 1"}$$k_{1}$ + $β_{2}$$1_{"category 2"}$$k_{2}$ + ... + $β_{n}$$1_{"category n"}$$k_{n}$ + $ε_{i}$ where $ε_{i}$ = N(0,$σ^{2}$) </center>
In this case, the model would demonstrate the contrast in the data between multiple categories. Each $β_{n}$ parameter would represent the contrast accounted for the data being a part of its respective category. In the case of several parameters, each parameter's associated $k_{n}$ would take a 1 if it is part of the category, or a 0 if not. All $k_{n}$ being a 0 would also represent another category, in which only the intercept has a non-zero value.

### Summary of Interaction

We discussed how a Multiple Linear Regression model behaves when it includes only indicator (dummy) variables derived from a non-binary categorical variable. This model form captures differences in the mean response of the dependent variable across different categories by creating binary variables for each category, except for one that serves as the reference group. The resulting linear form shows the mean effect of each category relative to the reference. The model estimates distinct mean outcomes for each category without implying any continuous trend, representing discrete shifts in the dependent variable between categories.

## 3. Use smf to fit multiple linear regression models to the course project dataset from the canadian social connection survey

```python
import pandas as pd
import statsmodels.formula.api as smf
import plotly.express as px
import numpy as np

np.random.seed(seed=7)

# Load the data
data = pd.read_csv("https://raw.githubusercontent.com/pointOfive/stat130chat130/main/CP/CSCS_data_anon.csv",
                   na_values=["9999", "", " ", "Presented but no response", "NA"])

# Filter to keep only rows where 'DATASET' is '2022 Cross-Sectional'
df = data[data['DATASET'] == '2022 Cross-Sectional']

df = df[df.REMOVE_case=='No'].copy()

# Select only the required columns
columns_to_keep = [
    'CONNECTION_social_num_close_friends',
    'LONELY_direct',
    'CONNECTION_reached_out_best_friend',
]
df = df[columns_to_keep]

# Remove rows with any missing data
df = df.dropna()

# Add a column 'is_lonely' based on the specified conditions
df['is_lonely'] = df['LONELY_direct'].apply(
    lambda x: 1 if x in ['All of the time (e.g. 5-7 days)]', 'Occasionally or a moderate amount of time (e.g. 3-4 days)'] else 0
)

# Add a column 'w_hang' based on the specified conditions
df['w_hang'] = df['CONNECTION_reached_out_best_friend'].apply(
    lambda x: 1 if x in ['Somewhat Likely', 'Likely', 'Very Likely'] else 0
)

# Step 1: Multiple Linear Regression (Additive specification)
mlr_model_additive = smf.ols(
    formula='is_lonely ~ CONNECTION_social_num_close_friends + w_hang',
    data=df
).fit()

# Print the summary of the additive MLR model
print("Additive MLR Model Summary")
print(mlr_model_additive.summary())

# Step 2: Multiple Linear Regression with a Synergistic (Interaction) specification
mlr_model_interaction = smf.ols(
    formula='is_lonely ~ CONNECTION_social_num_close_friends * w_hang',
    data=df
).fit()

# Print the summary of the interaction MLR model
print("\nInteraction MLR Model Summary")
print(mlr_model_interaction.summary())

# Create x_vals for plotting
x_vals = np.linspace(df['CONNECTION_social_num_close_friends'].min(),
                     df['CONNECTION_social_num_close_friends'].max(), 100)

# Add random noise to simulate real data
noise = np.random.normal(0, 0.1, size=x_vals.shape)

# Plot for the additive model (no interaction, just number of friends)
y_vals_additive = mlr_model_additive.predict(pd.DataFrame({
    'CONNECTION_social_num_close_friends': x_vals,
    'w_hang': 1  # Using w_hang = 1 as a reference
})) + noise

# Create the additive model plot (with a single line of best fit)
fig_additive = px.scatter(
    x=x_vals, 
    y=y_vals_additive, 
    labels={'x': 'Number of Close Friends', 'y': 'Predicted Loneliness'},
    title='Additive Model: Number of Friends vs Loneliness'
)

# Add only one line of best fit for the additive model (using w_hang = 1 as reference)
fig_additive.add_scatter(
    x=df['CONNECTION_social_num_close_friends'], 
    y=mlr_model_additive.predict(df.assign(w_hang=1)), 
    mode='lines', 
    name='Best Fit Line', 
    line=dict(color='red', dash='solid')
)

# Show the additive model plot
fig_additive.show(renderer='png')

# For the additive model
print("Additive Model Equation:")
additive_intercept = mlr_model_additive.params['Intercept']
additive_coeff_friends = mlr_model_additive.params['CONNECTION_social_num_close_friends']
additive_coeff_hang = mlr_model_additive.params['w_hang']
print(f"Equation: Predicted Loneliness = {additive_intercept:.4f} + ({additive_coeff_friends:.4f} * Number of Friends) + ({additive_coeff_hang:.4f} * w_hang)")

# Let's simulate the fitted line for the interaction model with varying w_hang values
x_vals = np.linspace(df['CONNECTION_social_num_close_friends'].min(),
                     df['CONNECTION_social_num_close_friends'].max(), 100)

# We will create predictions for two different values of w_hang: 0 and 1
y_vals_interaction_0 = mlr_model_interaction.predict(pd.DataFrame({
    'CONNECTION_social_num_close_friends': x_vals,
    'w_hang': 0  # For w_hang = 0
}))

y_vals_interaction_1 = mlr_model_interaction.predict(pd.DataFrame({
    'CONNECTION_social_num_close_friends': x_vals,
    'w_hang': 1  # For w_hang = 1
}))

# Add random noise to simulate real data
noise_0 = np.random.normal(0, 0.1, size=x_vals.shape)
noise_1 = np.random.normal(0, 0.1, size=x_vals.shape)

# Plot the interaction model's best fit line for both w_hang = 0 and w_hang = 1
fig_interaction = px.scatter(x=x_vals, y=y_vals_interaction_1 + noise_1, 
                             labels={'x': 'Number of Close Friends', 'y': 'Predicted Loneliness'},
                             title='Interaction Model: Number of Friends vs Loneliness (w_hang = 1)')
# Add lines of best fit for both w_hang = 0 and w_hang = 1
fig_interaction.add_scatter(x=df['CONNECTION_social_num_close_friends'], 
                            y=mlr_model_interaction.predict(df.assign(w_hang=1)), mode='lines', 
                            name='Best Fit Line (w_hang=1)', line=dict(color='red', dash='solid'))
fig_interaction.add_scatter(x=df['CONNECTION_social_num_close_friends'], 
                            y=mlr_model_interaction.predict(df.assign(w_hang=0)), mode='lines', 
                            name='Best Fit Line (w_hang=0)', line=dict(color='blue', dash='solid'))

fig_interaction.show(renderer='png')

print("\nInteraction Model Equation:")
interaction_intercept = mlr_model_interaction.params['Intercept']
interaction_coeff_friends = mlr_model_interaction.params['CONNECTION_social_num_close_friends']
interaction_coeff_hang = mlr_model_interaction.params['w_hang']
interaction_coeff_interaction = mlr_model_interaction.params['CONNECTION_social_num_close_friends:w_hang']
print(f"Equation: Predicted Loneliness = {interaction_intercept:.4f} + ({interaction_coeff_friends:.4f} * Number of Friends) + ({interaction_coeff_hang:.4f} * w_hang) + ({interaction_coeff_interaction:.4f} * (Number of Friends * w_hang))")
```

```python
import plotly.express as px
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from sklearn.model_selection import train_test_split
import plotly.graph_objects as go

url = "https://raw.githubusercontent.com/KeithGalli/pandas/master/pokemon_data.csv"
# fail https://github.com/KeithGalli/pandas/blob/master/pokemon_data.csv
pokemon_data = pd.read_csv(url) 

# Split data into training and testing
pokemon_training_data, pokemon_testing_data = train_test_split(pokemon_data, train_size=0.5)

# Clean the column names by replacing periods with underscores
pokemon_data.columns = pokemon_data.columns.str.replace('. ', '_')

# Now you can use the cleaned column names in the formula
linear_form = 'Speed ~ Attack * Defense + HP * Sp_Atk + Sp_Def * Sp_Atk + Sp_Def * Attack'


# Set number of repetitions for cross-validation
reps = 100

# Initialize arrays to store Rsquared values
in_sample_Rsquared = np.array([0.0]*reps)
out_of_sample_Rsquared = np.array([0.0]*reps)

for i in range(reps):
    # Split data into training and testing in each iteration
    pokemon_training_data, pokemon_testing_data = train_test_split(pokemon_data, train_size=0.5)
    
    # Fit the model to the training data
    final_model_fit = smf.ols(formula=linear_form, data=pokemon_training_data).fit()
    
    # Store in-sample Rsquared
    in_sample_Rsquared[i] = final_model_fit.rsquared
    
    # Calculate out-of-sample Rsquared using the testing data
    out_of_sample_Rsquared[i] = np.corrcoef(pokemon_testing_data.Speed, 
                                            final_model_fit.predict(pokemon_testing_data))[0,1]**2

# Create a DataFrame to store the results
df = pd.DataFrame({"In Sample Performance (Rsquared)": in_sample_Rsquared,
                   "Out of Sample Performance (Rsquared)": out_of_sample_Rsquared})

# Visualize the results with a scatter plot
fig = px.scatter(df, x="In Sample Performance (Rsquared)", 
                     y="Out of Sample Performance (Rsquared)")
fig.add_trace(go.Scatter(x=[0,1], y=[0,1], name="y=x", line_shape='linear'))

# Show the plot
fig.show(renderer="png")
```
